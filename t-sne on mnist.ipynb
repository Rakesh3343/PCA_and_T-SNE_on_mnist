```python
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
import seaborn as sns
```

```python
tsne=TSNE(n_components=2,perplexity=150)
```

```python
df=pd.read_csv(r"C:\Users\rakes\OneDrive\Desktop\course stuff\pca_on_MNIST\pca_mnist\mnist_csv.csv")
```

```python
mn_tsne=tsne.fit_transform(df.iloc[:,:784])
```

```python
mn_tsne.shape
```

```python
mn_np= np.vstack((mn_tsne.T,df['target'].values)).T
```

```python
final=pd.DataFrame(mn_np,columns=['f1','f2','target'])
```

```python
final['target']=final['target'].apply(lambda x:int(x))
```

```python
sns.scatterplot(x=final['f1'],y=final['f2'],hue=final['target'],palette='bright')
```

## *in higher dimensional space we can see they are forming nice clusters.*
